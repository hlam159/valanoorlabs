```python
from vlabs.spm import *
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
%matplotlib notebook
data = AsylumDART(os.path.join("data", "P0125OP0001.h5"))
pos = []
def onclick(event):
    pos.append([event.xdata/512 * ,event.ydata])

    xdata = event.xdata/512
    ydata = event.ydata/512
    #print(xdata)
    #print(ydata)
    ax.plot(xdata, ydata, "o")

fig, ax = data.single_image_plot("Topography")

fig.canvas.mpl_connect('button_press_event', onclick)
```
