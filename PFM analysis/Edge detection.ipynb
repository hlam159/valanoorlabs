```python
from vlabs.spm import *
import matplotlib.pyplot as plt
import numpy as np
scan1 = AsylumDART("data/P0125OP0004.h5")
```

```python

plt.style.use("standard")
fig, ax = scan1.single_image_plot("Topography")
#fig, ax = plt.subplots()
import skimage.feature

edges = skimage.feature.canny(
    image=scan1.topography,
    sigma=1,
    low_threshold=-1e-9,
    high_threshold=1e-9,
)
d = np.ma.masked_where(edges == False,  np.zeros((512,512)))
ax.imshow(d)
```
